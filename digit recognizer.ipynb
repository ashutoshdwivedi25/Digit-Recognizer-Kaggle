```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv(r"C:\Users\DaRK LoRD\Downloads\train (2).csv")
train.head()
```

```python
test = pd.read_csv(r"C:\Users\DaRK LoRD\Downloads\test (1).csv")
test.head()
```

```python
train_data = train.drop('label', axis=1)
target = train['label']
```

```python
train_data = np.array(train_data)
target = np.array(target)
```

```python
test = np.array(test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
clf = KNeighborsClassifier(n_neighbors = 13)
clf.fit(train_data, target)
prediction = clf.predict(test)
```

```python
count = []
for df in range (0,28000):
    count.append(df+1)
```

```python
print(count[0])
```

```python
submission = pd.DataFrame({
        "ImageId": count,
        "Label": prediction
    })

submission.to_csv('submission.csv', index=False)
```
